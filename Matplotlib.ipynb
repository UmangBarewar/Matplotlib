```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot();
```

```python
x= [1,2,3,4,5]
y=[11,22,33,44,55]
```

```python
plt.plot(x,y)
```

```python
# First method
fig = plt.figure() #creates a figure
ax = fig.add_subplot() # adds some axes
plt.show();
```

```python
#Second Method
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)#add some data
plt.show()
```

```python
# Third method (recommended)
fig, ax = plt.subplots()
ax.plot(x,y);
plt.show()
```

```python
# Day2
#1. Prepare Data
x= [1,2,3,4,5]
y=[11,22,33,44,55]

#2. Setup Plot
fig,ax = plt.subplots(figsize=(10,10))

#3.Plot data
ax.plot(x,y)

#4. Customize plot
ax.set(title='Simple PLto',xlabel="X-axis",ylabel="y-axis")

#5.Save ad show the whole figure
fig.savefig("Samplefig.png")

```

```python
#make a plot of dictionary
nut_butter_prices =  {"Almond butter ": 10,
                     "Peanut butter ": 8,
                     "Cashew butter ":12}
fig,ax=plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Nut butter store",ylabel="Price",xlabel="Butter");
```

```python
fig,ax=plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
ax.set(title="Nut butter store",ylabel="Price",xlabel="Butter");
```

## two options for subplots

```python
#1st option
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,4))
# plot to each differnet axes
ax1.plot(x,x/2);
ax2.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax3.scatter(np.random.random(10),np.random.random(10));
ax4.hist(np.random.random(10));
```

```python
#2nd option
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,4))
# plot to each differnet axes
ax[0,0].plot(x,x/2);
ax[0,1].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,0].scatter(np.random.random(10),np.random.random(10));
ax[1,1].hist(np.random.random(10));
```
